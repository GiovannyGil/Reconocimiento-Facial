```python
import cv2
import mediapipe as mp
```

```python
mp_face_mesh = mp.solutions.face_mesh
mp_drawing = mp.solutions.drawing_utils

# lista para los puntos de interes a acceder
index_list = [70, 63, 105, 66, 107, 336, 296, 334, 293, 300, 122, 196, 3, 51, 281, 248, 419, 351, 37, 0, 267,4, ]
```

```python
with mp_face_mesh.FaceMesh(
    static_image_mode=True, # True para imagenes fotos, false para video, stream
    max_num_faces=10, # cantidad de personas a las que le pondra la malla facial
    min_detection_confidence=0.5) as face_mesh:
    
    image = cv2.imread('./fotos-Prueba/personas2.jpg')  # Lee la imagen
    height, width, _ = image.shape # tamao
    #darle una imagen en RGB
    image_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    #aplicar deteccion
    results = face_mesh.process(image_rgb)
    
    print('Face Landmarks: ', results.multi_face_landmarks) 
    
    if results.multi_face_landmarks is not None: # asegurarse que si hay imagenes detectadas
        for face_landmarks in results.multi_face_landmarks: # lista de puntos del rostro detectado en cada iteracion
            
            # dar color a la malla facial
            """ mp_drawing.draw_landmarks(image, face_landmarks, mp_face_mesh.FACE_CONNECTIONS, mp_drawing.DrawingSpec(color=(0,0,0), thickness=1, circle_radius=1), mp_drawing.DrawingSpec(color=(0,0,0), thickness=1)) # ver los puntos y conecciones en la imagen con colores modificados """
            
            # acceder a un punto de interes -> nariz
            """ print(int(face_landmarks.landmark[4].x * width))
             x =  int(face_landmarks.landmark[4].x * width)
            y = int(face_landmarks.landmark[4].y * height)
            cv2.circle(image, (x,y),2,(255,0,255),2) """
            
            # acceder a los puntos de interes index_list
            for index in index_list:
                x = int(face_landmarks.landmark[index].x * width)
                y = int(face_landmarks.landmark[index].y * height)
                cv2.circle(image, (x,y),2, (255,255,255),1)
                
            
            
    
    
    cv2.imshow('Image', image)  # Muestra la imagen
    cv2.waitKey(0)  # Espera a que se presione una tecla
cv2.destroyAllWindows()
```
